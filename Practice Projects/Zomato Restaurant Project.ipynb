```python
# We want to merge the two dataframes on the country code column and keep the df table (left)
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
```

#### Distribution of votes

```python
# Clean up column names
final_df.columns = final_df.columns.str.strip().str.lower()

# Drop NaN values in "votes" column if necessary
final_df = final_df.dropna(subset=['votes'])

# Create the plot
plt.figure(figsize=(12, 8))
sns.displot(final_df, x="votes", kind="kde", color='blue')
plt.show()
```

#### Distribution of price range

```python
sns.displot(final_df, x="price range", kind="kde", multiple="stack",color='red')
plt.show()
```

```python
sns.displot(final_df, x="aggregate rating", kind="hist", multiple="stack",color='orange')
plt.show()
```

Observation:

```python
sns.displot(final_df, x="average cost for two", kind="kde", multiple="stack",color='green')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="has online delivery",data=final_df)
plt.title("RESTAURANT HAS ONLINE DELIVERY")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="has table booking",data=final_df)
plt.title("RESTAURANT HAS TABLE BOOKING")
plt.show()
```

Observation :
Maximum number of restaurants has no 'Online delivery' and 'Table booking' options

```python
cuisines =final_df.groupby("price range")["cuisines"].count().reset_index()
sns.barplot(data=cuisines,x="price range",y="cuisines",palette=['#a1c9f4', '#8de5a1', '#ff9f9b', '#d0bbff'])
plt.show()
```

Observation :
Maximum number of cuisines are in the price of 1

```python
vote=final_df.groupby("restaurant name")["votes"].sum().reset_index()
vote=vote.sort_values(by="votes")
vote=vote.tail(10)

plt.figure(figsize=(8,6))
sns.barplot(x="votes",y="restaurant name",data=vote,palette=("Spectral"))
plt.title("TOP 10 RESTAURANTS BASED ON VOTES")
plt.show()
```

Observation :
    
Barbeque Nation has more number of votes

Absolute Barbecues has second highest number of votes

```python
#costliest cuisines
acft=final_df[final_df['average cost for two']>200000]
acft=acft.sort_values(by='average cost for two')
sns.barplot(data=acft,x="cuisines",y="average cost for two",palette="YlOrBr")
plt.xticks(rotation=90)
plt.show()
```

Observation :
The above plot shows " Costliest cuisines "

```python
restaurant_counts=final_df.groupby('price range')['restaurant name'].count().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(x="price range",y="restaurant name",data=restaurant_counts,palette=("dark:#5A9_r"))
plt.title("RESTAURANTS COUNT BASED ON PRICE RANGE")
plt.ylabel('Restaurant Count')
plt.show()
```

Observation :
The above plot shows how many number of Restaurants are in the price of 1,2,3,4 .

4000+ Restaurants are in price range of one.
3000 Restaurants are in the price range of 2

```python
restaurant_counts=final_df.groupby('price range')['cuisines'].count().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(x="price range",y="cuisines",data=restaurant_counts,palette=("dark:#5A9_r"))
plt.title("CUISINES COUNT BASED ON PRICE RANGE")
plt.ylabel('Cuisines Count')
plt.show()
```

Observation :
The above plot shows how many number of Cuisines are in the price of 1,2,3,4 .

4000+ Cuisines are in the price range of "1"
