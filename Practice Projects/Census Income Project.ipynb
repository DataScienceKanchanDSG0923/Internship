## Data Preprocessing & Feature Engineering :

Encoding Categorical Variables :

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # label encoder 
df['Income']=le.fit_transform(df['Income']) 
df['Sex']=le.fit_transform(df['Sex'])
#Converting 2 columns into binary 
df = pd.get_dummies(df,drop_first=True)
pd.set_option('display.max_columns',100)#to display all columns
```

```python
df.head()
```

Now our almost data values is 0 and 1 except few features like “‘Age’,’Fnlwgt’,’Education_num’,’Hours_per_week’” we can use standard scaler we and convert those features in same scale.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_col_sacle = df[['Age','Fnlwgt','Education_num','Hours_per_week']]
train_scaler_col = scaler.fit_transform(train_col_sacle)
train_scaler_col = pd.DataFrame(train_scaler_col,columns=train_col_sacle.columns)
df['Age']= train_scaler_col['Age']
df['Fnlwgt']= train_scaler_col['Fnlwgt']
df['Education_num']= train_scaler_col['Education_num']
df['Hours_per_week']= train_scaler_col['Hours_per_week']
```

## Model Building

```python
X = df.drop(['Income'],axis=1)
y = df['Income']
```

### train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X,y, test_size=0.30, random_state=11)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize logistic regression model
lr = LogisticRegression()

# Drop rows with missing values in training data
X_train_clean = X_train.dropna()
y_train_clean = y_train[X_train_clean.index]

# Fit logistic regression model on the cleaned training data
lr.fit(X_train_clean, y_train_clean)

# Print the training score
print("Logistic Regression training score:", lr.score(X_train_clean, y_train_clean))

# Now, make predictions on the test set (make sure X_test has no missing values as well)
X_test_clean = X_test.dropna()
y_test_clean = y_test[X_test_clean.index]
y_pred = lr.predict(X_test_clean)

# Evaluate the model
print("Accuracy on test set:", accuracy_score(y_test_clean, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_clean, y_pred))
print("Classification Report:\n", classification_report(y_test_clean, y_pred))
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier() #KNearest Neibour

knn.fit(X_train_clean, y_train_clean)
print("knn classification score",knn.score(X_train_clean,y_train_clean))

y_pred = knn.predict(X_test_clean)

# Evaluate the model
print("Accuracy on test set:", accuracy_score(y_test_clean, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_clean, y_pred))
print("Classification Report:\n", classification_report(y_test_clean, y_pred))
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier() # Deciesion Tree

dt.fit(X_train_clean, y_train_clean)
print("dt classification score",dt.score(X_train_clean,y_train_clean))

y_pred = dt.predict(X_test_clean)

# Evaluate the model
print("Accuracy on test set:", accuracy_score(y_test_clean, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_clean, y_pred))
print("Classification Report:\n", classification_report(y_test_clean, y_pred))
```

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier() # Random Forest

rf.fit(X_train_clean, y_train_clean)
print("rf classification score",rf.score(X_train_clean,y_train_clean))

y_pred = rf.predict(X_test_clean)

# Evaluate the model
print("Accuracy on test set:", accuracy_score(y_test_clean, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_clean, y_pred))
print("Classification Report:\n", classification_report(y_test_clean, y_pred))
```

### GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gdboost=GradientBoostingClassifier() #Gradient Boosting Classifier

gdboost.fit(X_train_clean, y_train_clean)
print("gdboost classification score",gdboost.score(X_train_clean,y_train_clean))

y_pred = rf.predict(X_test_clean)

# Evaluate the model
print("Accuracy on test set:", accuracy_score(y_test_clean, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_clean, y_pred))
print("Classification Report:\n", classification_report(y_test_clean, y_pred))
```

```python
from xgboost import XGBClassifier
xgboost=XGBClassifier()  #Xtrim Gredient Boosting Classifier

xgboost.fit(X_train_clean, y_train_clean)
print("xgboost classification score",xgboost.score(X_train_clean,y_train_clean))

y_pred = rf.predict(X_test_clean)

# Evaluate the model
print("Accuracy on test set:", accuracy_score(y_test_clean, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_clean, y_pred))
print("Classification Report:\n", classification_report(y_test_clean, y_pred))
```

As our data was imbalanced and in this case we have to consider F1-score and there are 3 models “rf ,xgboost, gdboost” giving the scores above 90% and rest of below 90%

## ROC AUC Curve:

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
#from sklearn.metrics import plot_roc_curve

print(roc_auc_score(y_test_clean,lr.predict(X_test_clean)))
print(roc_auc_score(y_test_clean,knn.predict(X_test_clean)))
print(roc_auc_score(y_test_clean,dt.predict(X_test_clean)))
print(roc_auc_score(y_test_clean,rf.predict(X_test_clean)))
print(roc_auc_score(y_test_clean,gdboost.predict(X_test_clean)))
print(roc_auc_score(y_test_clean,xgboost.predict(X_test_clean)))
```

## HyperParameter Tuning

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

# Define your XGBoost model
xgboost_model = xgb.XGBClassifier()

# Define the parameter grid
parm_grid = {
    'max_depth': [3, 4],
    'subsample': [0.5, 0.8],
    'learning_rate': [0.1],
    'min_child_weight': [1, 2],
    'random_state': [4, 5]
}

# Set up GridSearchCV
gridsearch = GridSearchCV(estimator=xgboost_model, param_grid=parm_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the GridSearchCV
gridsearch.fit(X_train_clean, y_train_clean)

# Print the best parameters and best score
#print("Best parameters found: ", gridsearch.best_params_)
#print("Best cross-validation score: ", gridsearch.best_score_)
```

```python
Final_model=XGBClassifier(estimator=xgboost_model, param_grid=parm_grid, cv=5, scoring='accuracy', n_jobs=-1)
Final_model.fit(X_train_clean,y_train_clean)
pred=Final_model.predict(X_test_clean)
acc=accuracy_score(y_test_clean,pred)
print(acc*100)
```

```python
from sklearn import metrics
fpr,tpr,thresholds=metrics.roc_curve(y_test_clean,pred)
roc_auc=metrics.auc(fpr,tpr)
display=metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=Final_model)
display.plot()
```
