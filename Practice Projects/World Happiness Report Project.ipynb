```python
sns.set_style('darkgrid')
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(10,9)
plt.rcParams['figure.facecolor']='#FFE5B4'
```

```python
#plot b/w happiness and GDP
plt.rcParams['figure.figsize']=(15,7)
plt.title("plot between happiness score and GDP")
sns.scatterplot(x="Happiness Score",y="Economy (GDP per Capita)",hue="Region",data=df)
```

This is a strong correlation! In context, this makes a lot of sense, because the data shows that the greater the GDP per Capita, the greater the happiness score. Given the effects that money give when people have it, of course people would be happy with more money. If I may say, the data suggests that money really does "buy" happiness.

```python
GDP_region=df.groupby("Region")["Economy (GDP per Capita)"].sum()
GDP_region
```

```python
GDP_region.plot.pie(autopct='%1.1f%%')
plt.title("GDP by region")
plt.ylabel(" ")
```

```python
plt.rcParams["figure.figsize"]=(15,7)
sns.scatterplot(x=df.Family,y=df["Happiness Score"],hue=df["Region"],s=200)
plt.legend(loc="upper left",fontsize='12')
plt.xlabel("Family")
plt.ylabel("Happiness Score")
```

A somewhat correlated graph, it suggests that family and happiness indeed go together. It is understandable that being around family can make people happier. However, given the many outliers, one cannot assume that happiness is always dependent upon family. Let us continue to see if other vairables that have a strong correlation with happiness!

```python
#Total Countries
total_countries=df.groupby("Region")[["Country"]].count()
print(total_countries)
```

```python
cor= df.corr('pearson')
f,ax=plt.subplots(figsize=(10,5))
sns.heatmap(cor,mask=np.zeros_like(cor,dtype=np.bool_),cmap="Blues",square=True,ax=ax)
```

```python
Corruption=df.groupby("Region")[["Trust (Government Corruption)"]].mean()
Corruption
```

```python
plt.rcParams['figure.figsize']=(12,8)
plt.title("Perception of corruption in various Regions")
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Corruption Index",fontsize=15)
plt.xticks(rotation=30,ha="right")
plt.bar(Corruption.index, Corruption['Trust (Government Corruption)'])
```

```python
data_ten_happiest=df.sort_values(by = 'Happiness Rank')
top_10=data_ten_happiest.head(10)
```

```python
data_ten_leastHappy=df.sort_values(by = 'Happiness Rank',ascending = False)
Bottom_10=data_ten_leastHappy.head(10)
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))
plt.tight_layout(pad=2)
xlabels=top_10.Country
axes[0].set_title("top 10 happiest countries Life Expectancy")
axes[0].set_xticklabels(xlabels,rotation=45,ha="right")
sns.barplot(x=top_10.Country,y=top_10["Health (Life Expectancy)"],ax=axes[0])
axes[0].set_xlabel("Country Name")
axes[0].set_ylabel("Life Expectancy")

xlabels=Bottom_10.Country
axes[1].set_title("Bottom 10 least happy countries Life Expectancy")
axes[1].set_xticklabels(xlabels,rotation=45,ha="right")
sns.barplot(x=Bottom_10.Country,y=Bottom_10["Health (Life Expectancy)"],ax=axes[1])
axes[1].set_xlabel("Country Name")
axes[1].set_ylabel("Life Expectancy")
```

```python
plt.rcParams["figure.figsize"]=(15,7)
sns.scatterplot(x=df["Health (Life Expectancy)"],y=df["Happiness Score"],hue=df["Region"],s=200)
plt.legend(loc="upper left",fontsize='12')
plt.xlabel("Life Expectancy")
plt.ylabel("Happiness Score")
```

This is more correlated. It is good to know that Happiness is dependent upon health! It can be concluded that healthy people can live life to it's extent which generally makes them happier.

```python
plt.rcParams["figure.figsize"]=(15,7)
sns.scatterplot(x=df.Generosity,y=df["Happiness Score"],hue=df["Region"],s=200)
plt.legend(loc="upper right",fontsize='12')
plt.xlabel("Generosity")
plt.ylabel("Happiness Score")
```

This very weak correlation is weaker than government corruption's correlation. In context of the data, it seems as though the more generous someone is, the less likeley they find happiness. This is clearly an issue. There must be ways that society can implement to promote this seemingly negative morale. You would think that being generous promotes happiness. Despite this thinking, the data shows otherwise. Let's fix this!

```python
plt.rcParams["figure.figsize"]=(15,7)
sns.scatterplot(x=df.Freedom,y=df["Happiness Score"],hue=df["Region"],s=200)
plt.legend(loc="upper left",fontsize='12')
plt.xlabel("Freedom")
plt.ylabel("Happiness Score")
```

 Weak correlation between Happiness and Freedom. This certainly tells us that happiness does not depend on one's freedom.

```python
country=df.sort_values(by="Trust (Government Corruption)").head(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title("Countries with most Perception of corruption")
plt.xlabel("Country",fontsize=13)
plt.ylabel("Corruption Index",fontsize=13)
plt.xticks(rotation=30,ha="right")
plt.bar(country.Country, country['Trust (Government Corruption)'])
```

```python
country=df.sort_values(by="Trust (Government Corruption)").tail(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title("Countries with most Perception of corruption")
plt.xlabel("Country",fontsize=13)
plt.ylabel("Corruption Index",fontsize=13)
plt.xticks(rotation=30,ha="right")
plt.bar(country.Country, country['Trust (Government Corruption)'])
```

```python
#corruption v/s happiness
plt.rcParams["figure.figsize"]=(15,7)
sns.scatterplot(x=df["Happiness Score"],y=df["Trust (Government Corruption)"],hue=df["Region"],s=200)
plt.legend(loc="upper left",fontsize='14')
plt.xlabel("Happiness Score")
plt.ylabel("Corruption")
```

This is a very weak correlation. Thankfully, this is great, because data shows that most people don't think that government corruption makes them happy. Less government corruption is clustered at a higher-end happiness score, which is a good thing.
