Rank 
B) Name

```python
#opening the hostelworld page on automated chrome browser
driver=webdriver.Chrome()
driver.get(" https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
#maximize chrome window
driver.maximize_window()
```

```python
Name=[]
Artist=[]
Upload=[]
Views=[]

video_details= driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr')
i=1
for video in video_details:
    Name_1= driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr['+str(i)+']/td[1]').text
    Name.append(Name_1)

    Artist_1= driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr['+str(i)+']/td[2]').text
    Artist.append(Artist_1)

    Upload_1= driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr['+str(i)+']/td[4]').text
    Upload.append(Upload_1)

    Views_1= driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr['+str(i)+']/td[3]').text
    Views.append(Views_1)
    
    i =i+1
    
df = pd.DataFrame({'Video Name':Name,'Artist':Artist,'Upload':Upload,'Views':Views})
df
```

6.Scrape the details of Highest selling novels. 
A) Book name 
B) Author name 
C) Volumes sold 
D) Publisher 
E) Genre 
 Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
#opening the amazon page on automated chrome browser
driver=webdriver.Chrome()
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")
#maximize chrome window
driver.maximize_window()
```

```python

Book_name=[]
Author_name=[]
Volumes_sold=[]
Publisher=[]
Genre=[]
total_rows = driver.find_elements(By.XPATH,"//div[@class='embed block']/table/tbody/tr")

for i in range(1,len(total_rows)+1):
    title= driver.find_element(By.XPATH,"//div[@class='embed block']/table/tbody/tr["+str(i)+"]/td[2]")
    Book_name.append(title.text)
    name= driver.find_element(By.XPATH,"//div[@class='embed block']/table/tbody/tr["+str(i)+"]/td[3]")
    Author_name.append(name.text)
    volumes= driver.find_element(By.XPATH,"//div[@class='embed block']/table/tbody/tr["+str(i)+"]/td[4]")
    Volumes_sold.append(volumes.text)
    publisher_name= driver.find_element(By.XPATH,"//div[@class='embed block']/table/tbody/tr["+str(i)+"]/td[5]")
    Publisher.append(publisher_name.text)
    Genre_1= driver.find_element(By.XPATH,"//div[@class='embed block']/table/tbody/tr["+str(i)+"]/td[6]")
    Genre.append(Genre_1.text)

df = pd.DataFrame({"Title":Book_name,"Author":Author_name,"Volume Sales":Volumes_sold,"Publisher":Publisher,"Genre":Genre})
df
```

8.Details of Datasets from UCI machine learning repositories. 
Url = https://archive.ics.uci.edu/ You 
have to find the following details: 
A) Dataset name 
B) Data type 
C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute G) Year 
 Note: - from the home page you have to go to the Show All Dataset page through code. 

```python
#opening the amazon page on automated chrome browser
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
#maximize chrome window
driver.maximize_window()
time.sleep(3)

Datasets= driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/header/nav/ul/li[1]/a").click()
time.sleep(3)

Expand_All= driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]").click()

Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
No_of_Instance=[]
No_of_attribute=[]
Year=[]

Name= driver.find_elements(By.XPATH,"//div[@class='relative col-span-8 sm:col-span-7']/h2/a")
for Data in Name:
    Dataset_Name.append(Data.text)

D_Type= driver.find_elements(By.XPATH,"//div[@class='relative col-span-8 sm:col-span-7']/div/div[2]/span")
for typ in D_Type:
    Data_Type.append(typ.text)

Given_task= driver.find_elements(By.XPATH,"//div[@class='relative col-span-8 sm:col-span-7']/div/div[1]/span")
for tsk in Given_task:
    Task.append(tsk.text)

A_Type= driver.find_elements(By.XPATH,"//table[@class='col-span-full my-2 table sm:col-start-2']/tbody/tr/td[2]")
for atr in A_Type:
    Attribute_Type.append(atr.text)

No_instance= driver.find_elements(By.XPATH,"//div[@class='relative col-span-8 sm:col-span-7']/div/div[3]/span")
for instance in No_instance:
    No_of_Instance.append(instance.text)

No_attribute= driver.find_elements(By.XPATH,"//div[@class='relative col-span-8 sm:col-span-7']/div/div[4]/span")
for attribute in No_attribute:
    No_of_attribute.append(attribute.text)

Given_year= driver.find_elements(By.XPATH,"//table[@class='col-span-full my-2 table sm:col-start-2']/tbody/tr/td[3]")
for yr in Given_year:
    Year.append(yr.text)

df=pd.DataFrame({"Dataset Name":Dataset_Name,"Data Type":Data_Type,"Task":Task,"Attribute Type":Attribute_Type,"No of Instance":No_of_Instance,"No of attribute":No_of_attribute,"Year":Year})
df
```
