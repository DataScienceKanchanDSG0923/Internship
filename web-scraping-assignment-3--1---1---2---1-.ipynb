1.Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars. 

```python
import selenium 
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.webdriver import ActionChains, Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import urllib
```

```python
#opening the amazon page on automated chrome browser
driver=webdriver.Chrome()
driver.get("https://www.amazon.in/")
#maximize chrome window
driver.maximize_window()

def search_product(product):
    product_input= driver.find_element(By.XPATH,'//*[@id="twotabsearchtextbox"]')
    product_input.send_keys(product)
    Search= driver.find_element(By.ID,'nav-search-submit-button')
    Search.click()

product = input("Please enter your product name ")
search_product(product)
```

2.In the above question, now scrape the following details of each product listed in first 3 pages of your search 
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
import pandas as pd

# Define XPath expressions for each element
xpaths = {
    "brand_name": '//span[@class="a-size-base-plus a-color-base"]',
    "product_name": '//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]/a',
    "price": '//span[@class="a-price"]',
    "return_exchange": '//div[@id="anonCarousel3"]/ol/li/div/span',
    "expected_delivery": '//span[contains(text(), "Get it as soon as") or contains(text(), "Arrives before") or contains(text(), "Get it by") or contains(text(), "Ships from") or contains(text(), "Get it by") or contains(text(), "FREE Delivery by") or contains(text(), "Get it by") or contains(text(), "Usually ships within")]',
    "availability": '//span[@class="a-size-medium a-color-success"] | //span[@class="a-size-medium a-color-error"] | //span[@class="a-size-medium a-color-price"]',
    "product_url": '//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]/a'
}

# Create an empty list to store product information
product_info = []

# Iterate through each XPath and scrape the corresponding information
for i in range(len(driver.find_elements(By.XPATH, xpaths["brand_name"]))):
    try:
        brand_name = driver.find_elements(By.XPATH, xpaths["brand_name"])[i].text
        product_name = driver.find_elements(By.XPATH, xpaths["product_name"])[i].text
        price = driver.find_elements(By.XPATH, xpaths["price"])[i].text
        return_exchange = driver.find_elements(By.XPATH, xpaths["return_exchange"])[i].text
    except:
        return_exchange = "-"
    try:
        expected_delivery = driver.find_elements(By.XPATH, xpaths["expected_delivery"])[i].text
    except:
        expected_delivery = "-"
    try:
        availability = driver.find_elements(By.XPATH, xpaths["availability"])[i].text
    except:
        availability = "-"
    try:
        product_url = driver.find_elements(By.XPATH, xpaths["product_url"])[i].get_attribute("href")
    except:
        product_url = "-"
    product_info.append((brand_name, product_name, price, return_exchange, expected_delivery, availability, product_url))

# Create a DataFrame from the product information
df = pd.DataFrame(product_info, columns=["Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Product URL"])

# Save the DataFrame to a CSV file
df.to_csv("product_info.csv", index=False)

# Print the DataFrame
print(df)
```

8.Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
from any YouTube Video. 

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get('https://www.youtube.com/watch?v=dx4Teh-nv3A')
driver.maximize_window()

```

```python
comments = []
upvotes = []
posted_times  = []

# Set to store unique comment IDs
comment_ids = set()
# Scroll down function
def scroll_down(driver, scroll_increment):
    driver.execute_script(f"window.scrollBy(0, {scroll_increment})")
    time.sleep(2)  # Adjust the wait time as needed

# Scroll down until enough comments are loaded
while True:
    # Find all comment elements
    comment_elements = driver.find_elements(By.XPATH, '//*[@id="content-text"]/span')
    upvote_elements = driver.find_elements(By.XPATH, '//span[@id="vote-count-middle"]')
    time_elements = driver.find_elements(By.XPATH, '//span[@id="published-time-text"]')
    
    # Store the text of each comment in the comments list
    for comment_element, upvote_element, time_element in zip(comment_elements, upvote_elements, time_elements):
        comment_id = comment_element.get_attribute("id")
        if comment_id not in comment_ids:
            upvotes.append(upvote_element.text)
            posted_times.append(time_element.text)
            comments.append(comment_element.text)
            
    # Check if enough comments are loaded
    if len(comments) >= 510:
        break
    
    # Scroll down to load more comments
    scroll_down(driver, 500)

# Print the scraped comments
for index, (comment, upvote, posted_time) in enumerate(zip(comments, upvotes, posted_times), start=1):
    print(f"Comment {index}: {comment} | Upvotes: {upvote} | Posted Time: {posted_time}")
```
