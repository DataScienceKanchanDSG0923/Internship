1.Write a python program to display IMDB’s Top rated 100 Indian movies’ data
https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_top_100_movies():
    url = "https://www.imdb.com/list/ls056092300/"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        movies = []
        for movie in soup.find_all("div", class_="lister-item-content"):
            title = movie.find('a').text.strip()
            rating = movie.find("span", class_="ipl-rating-star__rating").text.strip()
            year = movie.find("span", class_="lister-item-year text-muted unbold").text.strip()[1:-1]
            movies.append({'Title': title, 'Rating': rating, 'Year': year})
        
        df = pd.DataFrame(movies)
        return df
    else:
        print("Failed to fetch the webpage.")
        return None

top_100_movies_df = get_top_100_movies()

print(top_100_movies_df)
```

6.Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_bewakoof_bestsellers(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        products = []

        for container in soup.find_all('div', class_='plp-product-card')[:10]:      
            name_tag = container.find('h2',class_='clr-shade4 h3-p-name undefined false')
            if name_tag:
                name=name_tag.text.strip()
            else:
                name="not available"
            image_url = container.find('img')['src']
            print(image_url)

            actual_price_element = container.find('div', class_='actualPriceText clr-shade5')
            if actual_price_element:
                actual_price = actual_price_element.text.strip()
            else:
                actual_price = "Price not available"
            discounted_price_element = container.find('div', class_='discountedPriceText clr-p-black false')
            if discounted_price_element:
                discounted_price = discounted_price_element.text.strip()
            else:
                discounted_price = "Price not available"          
            products.append({'Product Name': name, 'Actual Price': actual_price,'Discounted Price':discounted_price,'Image URL': image_url})

        df = pd.DataFrame(products)
        return df
    else:
        print("Failed to retrieve page.")
        return None
url = "https://www.bewakoof.com/bestseller?sort=popular"
bestsellers_df = scrape_bewakoof_bestsellers(url)
print(bestsellers_df)                  
```

7.Please visit https://www.cnbc.com/world/?region=world and scrapa) headings
b) date
c) News link

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def scrape_cnbc_world_news():
    url = "https://www.cnbc.com/world/?region=world"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")     
        articles = soup.find_all("div", class_="LatestNews-headlineWrapper")

        for article in articles:
            headline = article.text.strip()  
            link = article.find("a", href=True)["href"]  
            absolute_link = link if "://" in link else urljoin(url, link)

            article_response = requests.get(absolute_link)
            if article_response.status_code == 200:
                article_soup = BeautifulSoup(article_response.content, "html.parser")
                date_element = article_soup.find("time")
                date = date_element["datetime"] if date_element else "No date found"
        
                print("Headline:", headline)
                print("Date:", date)
                print("Link:", absolute_link)    
                print("-" * 50)
    else:
        print("Failed to fetch the page")

scrape_cnbc_world_news()
```

8.Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/ and scrap-
 a) Paper title
 b) date
 c) Author

```python
import requests
from bs4 import BeautifulSoup

def scrape_most_downloaded_articles(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    articles = []
    article_blocks = soup.find_all('div', class_='article-listing')
    for block in article_blocks:
        title = block.find('h2').text.strip()
        date = block.find('p', class_='article-date').text.strip()
        author = block.find('p', class_='article-authors').text.strip()
        
        article_data = {
            'Title': title,
            'Date': date,
            'Author': author
        }
        articles.append(article_data)

    return articles

url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
most_downloaded_articles = scrape_most_downloaded_articles(url)

for article in most_downloaded_articles:
    print("Title:", article['Title'])
    print("Date:", article['Date'])
    print("Author:", article['Author'])
    print()
```
